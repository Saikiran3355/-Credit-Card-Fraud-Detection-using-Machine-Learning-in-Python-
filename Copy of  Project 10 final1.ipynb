Build a sample dataset containing similar distributions  of normal transactions and Fraudulent Transaction

Number of fraudulent Transaction --> 492

```python
import pandas as pd
import numpy as np
legit_data = pd.DataFrame({
    'feature1' : np.random.rand(1000),
    'feature2' : np.random.rand(1000),
})
```

```python
import pandas as pd
import numpy as np


legit_sample = pd.DataFrame({
    'feature1': np.random.rand(1000),
    'feature2': np.random.rand(1000),
    'Class': 0  # Add 'Class' column with label 0 for legit transactions
})

fraud = pd.DataFrame({
    'feature1': np.random.normal(0.5, 0.1, 50),  # Different distribution
    'feature2': np.random.normal(0.5, 0.1, 50),
    'Class': 1  # Add 'Class' column with label 1 for fraudulent transactions
})


new_dataset = pd.concat([legit_sample, fraud], axis=0)


print("Dataset Shape:", new_dataset.shape)
print("\nFirst 5 rows:")
print(new_dataset.head())
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data in to Features & Target

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data in to Training data & Testing Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model Event

Accuracy Score

```python
from sklearn.metrics import accuracy_score
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
